```python
#to count the number of patient for each category of the chest pains
ChestPain = df.ChestPain.values
typ = 0
ageTyp = []
asymp = 0
ageAsymp = []
nonang = 0
ageNonang = []
nontyp = 0
ageNontyp = []
for types in ChestPain:
    if types == "typical":
        typ += 1
        
    elif types == "asymptomatic":
        asymp += 1
        
    elif types == "nonanginal":
        nonang += 1
        
    else:
        nontyp += 1

print ("The number of patients that have typical chest pain is {}". format(typ))
print ("The number of patients that have asymptomatic chest pain is {}". format(asymp))
print ("The number of patients that have nonanginal chest pain is {}". format(nonang))
print ("The number of patients that have nontypical chest pain is {}". format(nontyp))
```

```python
#to print out ages
ChestPain = df.ChestPain.values
Age = df.Age.values
ageTyp = []
ageAsymp = []
ageNonang = []
ageNontyp = []
for types, age in zip(ChestPain, Age):
    if types == "typical":
        ageTyp.append(age)
        
    elif types == "asymptomatic":
        ageAsymp.append(age)
        
    elif types == "nonanginal":
        ageNonang.append(age)
        
    else:
        ageNontyp.append(age)
        
print ("The ages of patients that have typical chest pain are {}". format(ageTyp))
print ("The ages of patients that have asymptomatic chest pain are {}". format(ageAsymp))
print ("The ages of patients that have nonanginal chest pain are {}". format(ageNonang))
print ("The ages of patients that have nontypical chest pain are {}". format(ageNontyp))
```

```python
#to get the average age for each category of chestpains
ageTypTable = pd.DataFrame(ageTyp)
ageTypTable["Typical Chest Pain"] = ageTypTable
a = ageTypTable["Typical Chest Pain"].mean()

ageAsympTable = pd.DataFrame(ageAsymp)
ageAsympTable["Asymptomatic  Chest Pain"] = ageAsympTable
b = ageAsympTable["Asymptomatic  Chest Pain"].mean()

ageNonagnTable = pd.DataFrame(ageNonang)
ageNonagnTable["nonaginal Chest Pain"] = ageNonagnTable
c= ageNonagnTable["nonaginal Chest Pain"].mean()

ageNontypTable = pd.DataFrame(ageNontyp)
ageNontypTable["nontypical Chest Pain"] = ageNontypTable
d = ageNontypTable["nontypical Chest Pain"].mean()


print ("The average age of patients that have typical chest pain is {}". format(a))
print ("The average age of patients that have asymptomatic chest pain is {}". format(b))
print ("The average of patients that have nonanginal chest pain is {}". format(c))
print ("The average of patients that have nontypical chest pain is {}". format(d))
```

```python
#to show the number of patients that have Sex = "1" and Sex = "0"
Sex = df.Sex.values
sex1 = 0
sex0 = 0
for i in Sex:
    if i == 1.0:
        sex1 += 1
    else:
        sex0 += 1
print("The number of patients that have sex = '1' equals: {}".format(sex1))
print("The number of patients that have sex = '0' equals: {}".format(sex0))
```
